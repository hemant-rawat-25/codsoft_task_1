```python
#required libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

# Loading and Processing Data

```python
titanic_data=pd.read_csv("tested.csv",encoding="unicode_escape")
titanic_data
```

```python
titanic_data.info()
```

```python
titanic_data.drop(["Ticket","Name","PassengerId"],axis=1,inplace=True)
titanic_data
```

```python
titanic_data["Sex"].replace({"male":0,"female":1},inplace=True)
titanic_data
```

```python
titanic_data["Embarked"].replace({'S':0,'C':1,'Q':2},inplace=True)
titanic_data
```

# Handing Missing Values

```python
#checking number of null values in each column
titanic_data.isnull().sum()
```

```python
titanic_data.drop(["Cabin"],axis=1,inplace=True)
titanic_data
```

```python
#fill missing age values with the average age
titanic_data["Age"].fillna(titanic_data["Age"].mean(),inplace=True)
```

```python
#fill missing fare values with the average fare
titanic_data["Fare"].fillna(titanic_data["Fare"].mean(),inplace=True)
```

```python
#fill missing embarked values with the most common embarked value
titanic_data["Embarked"].fillna(titanic_data["Embarked"].mode(),inplace=True)
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data
```

# Splitting the Data

```python
#spliting the data into the feature x and target variable y
X=titanic_data.drop(["Survived"],axis=1)
Y=titanic_data["Survived"]
```

```python
#splitting the dataset into tarining and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=4)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Training Model

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train,Y_train)
```

# Model Evaluation

```python
#accuracy on traing data 
X_train_prediction=model.predict(X_train)
print(X_train_prediction)
```

```python
training_data_accurancy=accuracy_score(Y_train,X_train_prediction)
print(training_data_accurancy)
```

```python
#acuuracy on test data
X_test_prediction=model.predict(X_test)
print(X_test_prediction)
```

```python
test_data_accurancy=accuracy_score(Y_test,X_test_prediction)
print("accuracy : " ,test_data_accurancy)
```
